# Extract Metadata from Images
This notebook extracts date/time metadata from images in the img/ folder using EXIF data.

```python
from PIL import Image, ExifTags
from datetime import datetime

image_path = 'img/IM_00001.JPG'
img = Image.open(image_path)
exif_data = img.getexif()

# Metadata
for tag_id, value in exif_data.items():
    tag_name = ExifTags.TAGS.get(tag_id, tag_id)
    print(f'{tag_name}: {value}')
```

```python
datetime_str = exif_data.get(306)

if datetime_str:
    dt = datetime.strptime(datetime_str, '%Y:%m:%d %H:%M:%S')
    date = dt.date()
    time = dt.time()
    
    print(f'Date: {date}')
    print(f'Time: {time}')
    print(f'\nFull datetime: {dt}')
else:
    print('No datetime found in EXIF')
```
